```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
file_path = r'C:\Users\Hager\fifa21_male2.csv'
df = pd.read_csv(file_path)
```

```python
#Age analysis
#Distribution of player ages.
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'], bins=20, kde=True)
plt.title('Distribution of Player Ages')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
fifa_position = df['Position'].unique()
```

```python
#Position Analysis
#the top 10 positions
plt.figure(figsize=(12, 6))
sns.countplot(x='Position', data=df, order=df['Position'].value_counts().head(30).index)
plt.title('Top 30 Player Positions')
plt.xlabel('Position')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
#Top players
top_players = df.sort_values(by='OVA', ascending=False).head(10)
print(top_players[['Name', 'OVA']])
```

```python
#Potential vs. Overall Rating
#Visualize the relationship between potential and overall rating:
plt.figure(figsize=(10, 6))
sns.scatterplot(x='POT', y='OVA', data=df)
plt.title('Potential vs Overall Rating')
plt.xlabel('Potential Rating')
plt.ylabel('Overall Rating')
plt.show()
```

```python
#Skill Analysis
#Explore the distribution of specific skill ratings.
#Identify players with exceptional skills in certain areas

skills_to_analyze = ['Dribbling', 'Finishing', 'Curve', 'FK Accuracy', 'Long Passing', 'Ball Control']
plt.figure(figsize=(16, 10))
for skill in skills_to_analyze:
    plt.subplot(2, 3, skills_to_analyze.index(skill) + 1)
    sns.histplot(df[skill], kde=True)
    plt.title(f'Distribution of {skill}')
    plt.xlabel(skill)
    plt.ylabel('Count')

plt.tight_layout()
plt.show()

# Identify top players in each skill
top_players_by_skill = df[['Name'] + skills_to_analyze].nlargest(5, columns=skills_to_analyze)
print("Top Players in Each Skill:")
print(top_players_by_skill)
```

```python
# Scatter plot for potential growth vs. age
plt.figure(figsize=(12, 6))
sns.scatterplot(x='Age', y='POT', data=df, alpha=0.5)
plt.title('Potential Growth vs. Age')
plt.xlabel('Age')
plt.ylabel('Potential Growth')
plt.show()

# Identify players with the highest potential growth
top_growth_players = df[['Name', 'Age', 'POT']].nlargest(5, columns='POT')
print("Players with the Highest Potential Growth:")
print(top_growth_players)
```

```python
import numpy as np


# List of physical attribute columns to analyze
physical_attributes = ['Height','Strength']

# Plot histograms for each physical attribute
plt.figure(figsize=(16, 6))
for i, attribute in enumerate(physical_attributes, 1):
    plt.subplot(1, 3, i)
    sns.histplot(df[attribute], kde=True, bins=30)  # Adjust the number of bins as needed
    plt.title(f'Distribution of {attribute}')
    plt.xlabel(attribute)
    plt.ylabel('Count')
    
    # Customize x-axis ticks
    plt.xticks(rotation=45)
    plt.tick_params(axis='x', which='both', bottom=False, top=False)  # Hide ticks on the bottom axis

plt.tight_layout()
plt.show()
```

```python
# Identify the top 10 values for Weight
top_10_weights = df['Weight'].value_counts().head(20).index

plt.figure(figsize=(10, 6))
sns.histplot(df[df['Weight'].isin(top_10_weights)]['Weight'], bins=10, kde=True)

plt.title('Distribution of Weight (Top 20)')
plt.xlabel('Weight')
plt.ylabel('Count')

# Customize x-axis ticks
plt.xticks(rotation=45)
plt.tick_params(axis='x', which='both', bottom=False, top=False)  # Hide ticks on the bottom axis

plt.show()
```

```python
# Distribution of Attacking Work Rate
plt.figure(figsize=(12, 6))
sns.countplot(x='A/W', data=df)
plt.title('Distribution of Attacking Work Rate')
plt.xlabel('Attacking Work Rate')
plt.ylabel('Count')
plt.show()
```

```python
# Distribution of Defensive Work Rate
plt.figure(figsize=(12, 6))
sns.countplot(x='D/W', data=df)
plt.title('Distribution of Defensive Work Rate')
plt.xlabel('Defensive Work Rate')
plt.ylabel('Count')
plt.show()
```

```python
# Impact of Defensive Work Rate on Overall Rating
plt.figure(figsize=(12, 6))
sns.boxplot(x='D/W', y='OVA', data=df)
plt.title('Impact of Defensive Work Rate on Overall Rating')
plt.xlabel('Defensive Work Rate')
plt.ylabel('Overall Rating')
plt.show()
```

```python
# Distribution of Players by Nationality (Top 10)
plt.figure(figsize=(16, 6))
sns.countplot(x='Nationality', data=df, order=df['Nationality'].value_counts().head(10).index)
plt.title('Top 10 Nationalities of Players')
plt.xlabel('Nationality')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Average Overall Rating by Nationality (Top 10)
avg_rating_by_nationality = df.groupby('Nationality')['OVA'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(16, 6))
sns.barplot(x=avg_rating_by_nationality.index, y=avg_rating_by_nationality.values)
plt.title('Average Overall Rating by Nationality (Top 10)')
plt.xlabel('Nationality')
plt.ylabel('Average Overall Rating')
plt.xticks(rotation=45)
plt.show()
```

```python
# Top Nationalities Represented in the Dataset (Top 10)
top_nationalities = df['Nationality'].value_counts().head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x=top_nationalities.index, y=top_nationalities.values)
plt.title('Top Nationalities Represented in the Dataset (Top 10)')
plt.xlabel('Nationality')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Distribution of Players Across Different Clubs (Top 10)
plt.figure(figsize=(16, 6))
sns.countplot(x='Club', data=df, order=df['Club'].value_counts().head(10).index)
plt.title('Top 10 Clubs with Most Players')
plt.xlabel('Club')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Average Overall Rating by Club (Top 10)
avg_rating_by_club = df.groupby('Club')['OVA'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(16, 6))
sns.barplot(x=avg_rating_by_club.index, y=avg_rating_by_club.values)
plt.title('Average Overall Rating by Club (Top 10)')
plt.xlabel('Club')
plt.ylabel('Average Overall Rating')
plt.xticks(rotation=45)
plt.show()
```

```python
# Top Clubs with the Highest-Rated Players (Top 10)
top_clubs_highest_rated = df.groupby('Club')['OVA'].max().sort_values(ascending=False).head(10)

plt.figure(figsize=(16, 6))
sns.barplot(x=top_clubs_highest_rated.index, y=top_clubs_highest_rated.values)
plt.title('Top 10 Clubs with Highest-Rated Players')
plt.xlabel('Club')
plt.ylabel('Highest Overall Rating')
plt.xticks(rotation=45)
plt.show()
```

```python
# Average Ratings for Different Playing Positions
# Calculate the average ratings for each playing position
avg_ratings_by_position = df.groupby('Position')['OVA'].mean().sort_values(ascending=False)

# Select the top 10 playing positions
top_10_positions = avg_ratings_by_position.head(10).index

# Filter the dataframe to include only the top 10 positions
df_top_10_positions = df[df['Position'].isin(top_10_positions)]

# Plot the average ratings for the top 10 playing positions
plt.figure(figsize=(16, 6))
sns.barplot(x='Position', y='OVA', data=df_top_10_positions, order=top_10_positions)
plt.title('Average Ratings for Top 10 Playing Positions')
plt.xlabel('Playing Position')
plt.ylabel('Average Overall Rating')
plt.xticks(rotation=45)
plt.show()
```

```python
# Top 10 Contract Lengths
top_10_contracts = df['Contract'].value_counts().head(10).index

# Contract Length Distribution (Top 10)
plt.figure(figsize=(12, 6))
sns.countplot(x='Contract', data=df[df['Contract'].isin(top_10_contracts)])
plt.title('Top 10 Contract Lengths Distribution')
plt.xlabel('Contract Length')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```
